### Example Starter Code


**Part 1 and 2: Using the starter code**
1.  Load the Dataset
*   Import the Titanic dataset directly using Seaborn's load_dataset function.

2. Perform Basic Data Exploration
*   Display the first few rows of the dataset.
*   Print the dataset's shape, columns, and summary information (info and describe methods).
*  Check for missing values in the dataset.

```python

# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
data = sns.load_dataset("titanic")

# Display First Few Rows
print(data.head())
```

```python
# Print Dataset Information
print("Print Dataset Information")
print(data.info())

#Shape of the date frame
print("Shape of the data frame:")
print(data.shape)

#checks the columns of the date frame
print("Columns of the data frame:")
print(data.columns)

# Print Dataset details
print("Print Dataset details")
print(data.describe())

# Check for Missing Values
print("Check for Missing Values")
print(data.isnull().sum())
```

**Part 3: Answer Key Questions with Visualizations**

Create visualizations to answer the following questions:

1.  How does the survival rate vary by passenger class (pclass)?
2.  What is the distribution of ages among passengers? Use a histogram or KDE plot.
3.  Are survival rates different between males and females?
4.  How does the survival rate differ for passengers with family aboard (sibsp, parch)?




```python
# 1. How does the survival rate vary by passenger class (pclass)?
sns.countplot(x='pclass', hue='survived', palette=['b', 'g'], data=data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# 2. What is the distribution of ages among passengers? Use a histogram or KDE plot.
sns.histplot(data['age'].dropna(), kde=True, color='skyblue')
plt.title('Distribution of Ages Among Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Are survival rates different between males and females?
sns.countplot(x='sex', hue='survived', palette=['b', 'g'], data=data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# 4. How does the survival rate differ for passengers with family aboard (sibsp, parch)?
data['family_size'] = data['sibsp'] + data['parch']+1
sns.countplot(x='family_size', hue='survived', palette=['b', 'g'], data=data)
plt.title('Survival Rate by Siblings/Spouses Aboard')
plt.show()
```

**Part 4: Bonus Tasks**

*   Explore relationships between multiple variables using a pair plot.
*   Create a heatmap to visualize any correlations in the dataset.

```python
sns.pairplot(data, hue ='survived')
plt.show()
```

```python
# Compute correlation matrix
co_mtx = data.corr(numeric_only=True)

# Print correlation matrix
print(co_mtx)

# Plot correlation heatmap
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)

# Display heatmap
plt.show()
```
